```python
import os
import shutil
import json

# File paths
testing_json_path = 'Testing_Set_Final.json'
validation_json_path = 'Validation_Set_Final.json'
source_dir = '/home/ali/projects/gully/labeling_tool/data/MO+IA_test_data_numbered'  # Replace with your source directory
output_dir = '/home/ali/projects/gully/GD_Images'  # Replace with your target directory

# Load JSON data
with open(testing_json_path, 'r') as f:
    testing_data = json.load(f)

with open(validation_json_path, 'r') as f:
    validation_data = json.load(f)

# Combine keys from both JSON files
tile_numbers = set(testing_data.keys()).union(validation_data.keys())

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Iterate through files in the source directory
for filename in os.listdir(source_dir):
    if filename.endswith('.tif'):
        # Extract the tile number from the filename
        parts = filename.split('_')
        tile_number = parts[-1].replace('tile_', '').replace('.tif', '')
        if tile_number in tile_numbers:
            # Modify the filename to remove the 'pos' or 'neg' prefix
            if parts[0] in ('pos', 'neg'):
                new_filename = '_'.join(parts[1:])
            else:
                new_filename = filename

            # Copy matching file to the output directory with the new name
            source_path = os.path.join(source_dir, filename)
            target_path = os.path.join(output_dir, new_filename)
            shutil.copy2(source_path, target_path)
            print(f"Copied: {filename} as {new_filename} to {output_dir}")
```
