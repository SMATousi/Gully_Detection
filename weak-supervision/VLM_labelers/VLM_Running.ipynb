```python
def check_yes_no(text):
    # Strip any leading/trailing whitespace and convert to lowercase
    text = text.strip().lower()

    # Check if the text starts with 'yes' or 'no'
    if text.startswith("yes"):
        return "4"
    elif text.startswith("no"):
        return "0"
    else:
        return "2" 
```
